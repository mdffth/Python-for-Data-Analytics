```python
penjualan = [200,300,250,400,350,275,325]
total_penjualan = sum(penjualan)
rata2 = total_penjualan / len(penjualan)

print(total_penjualan)
print(rata2)

if total_penjualan>2000:
  if rata2>300:
    print("Penjualan bagus hari ini, rata rata penjualan memuaskan")
  else:
    print("Penjualan bagus minggu ini")
else:
  print("Perlu meningkatkan penjualan minggu depan")
```

```python
jumlah_barang = 110
berat_barang = 3.5
barang_baru = True

#Barang memenuhi persyaratan (AND)
if jumlah_barang > 100 and berat_barang > 3:
  print("Persediaan barang memenuhi persyaratan")
else:
  print("Persediaan kurang")

#Barang prioritas (OR)
if not barang_baru or berat_barang < 4:
  print("Barang ini prioritas")
else:
  print("Barang ini bukan prioritas")

#Barang ekspor

if (jumlah_barang > 120) or (barang_baru and berat_barang > 3):
  print("Barang layak ekspor")
else:
  print("Barang tidak layak ekspor")

#Barang tidak sesuai untuk di audit

if jumlah_barang <= 100 and not barang_baru:
  print("Barang tidak sesuai untuk di audit")
else:
  print("Barang sesuai di audit")

#Barang ideal untuk pengiriman lokal
if (jumlah_barang > 50 and (berat_barang >= 2 and berat_barang<=5) and barang_baru):
  print("Barang ideal untuk pengiriman lokal")
else:
  print("Barang tidak ideal untuk pengiriman lokal")
```

```python
# Membuat array 1D, 2D, 3D

#Array 1D
list_1D = [1,2,3,4,5]
array_1D = np.array(list_1D)
print(f"Array 1D : \n {array_1D}")

#Array 2D
list_2D = [[1,2],[3,4]]
array_2D = np.array(list_2D)
print(f"Array 2D : \n {array_2D}")

#Array 3D
list_3D = [[[1,2,3]],[[4,5,6]],[[7,8,9]]]
array_3D = np.array(list_3D)
print(f"Array 3D : \n {array_3D}")
```

```python
#membuat array dari range
# variabel = arange(start, stop, step ,type)

#membuat array dengan batas 10
array1 = np.arange(10)
print(f"Array 0-9 : {array1}")

#membuat array dari 1 - 20 dengan 2 step
array2 = np.arange(1,21,2)
print(f"Array 1-19 step 2 : {array2}")
```

```python
#membuat array dari zeros dan ones
# zeros adalah array  yang berisi 0 saja
# ones adalah array  yang berisi 1 saja

# variable = np.zeros((shape),dtype)

#array 2D

array1D = np.zeros((2,3))
print(f"Array 2D : {array1D}")

#array 3D type: int
array3D = np.ones((2,2,3), dtype=int)
print(f"Array 3D : {array3D}")
```

```python
#Mengatur dimensi array (reshape, flatten)

#reshape mengganti dimensi array
#flatten meratakan ke 1dimensi array

arr1 = np.arange(12)
print(f"Array 1D : {arr1}")

#mengganti menjadi 2d

arr2 = arr1.reshape(3,4)
print(f"Array 2D : {arr2}")

#mengganti menjadi 3d

arr3 = arr2.reshape(2,2,3)
print(f"Array 3D : {arr3}")

array_flatten = arr3.flatten()
print(f"Array yg sdh diratakan : {array_flatten}")
```

```python
# distribusi normal dan uniform
# digunakan gambaran persebaran data
# distribusi normal : ditribusi yang memiliki persebaran data simetris | variabe = np.random.normal(mean, sclae, size )
import numpy as np
import matplotlib.pyplot as plt

# Set seed untuk hasil tetap
np.random.seed(42)

# Membuat data distribusi normal
dis_normal = np.random.normal(loc=6, scale=6, size=10000)
dis_sorted = np.sort(dis_normal)

# Plot
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Grafik 1: Data asli
axs[0].bar(range(len(dis_normal)), dis_normal, color='skyblue')
axs[0].set_title("Data Asli (Tanpa Diurutkan)")
axs[0].set_xlabel("Indeks")
axs[0].set_ylabel("Nilai")

# Grafik 2: Data setelah di-sort
axs[1].plot(dis_sorted, marker='o', linestyle='-', color='green')
axs[1].set_title("Data Setelah Diurutkan")
axs[1].set_xlabel("Urutan (setelah sort)")
axs[1].set_ylabel("Nilai")

plt.tight_layout()
plt.show()
print(dis_normal.mean())

# distribusi uniform : ditribusi yang memiliki persebaran data simetris | variabe = np.random.normal(low, high, size )
dis_uniform = np.random.uniform(1, 10000, size=10000)
# Distribusi uniform: nilai antara 1 dan 6, ukuran 10000
dis_uniform_sorted = np.sort(dis_uniform)

# Plot uniform
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Grafik 1: Data uniform asli
axs[0].bar(range(len(dis_uniform)), dis_uniform, color='orange')
axs[0].set_title("Distribusi Uniform (Tanpa Diurutkan)")
axs[0].set_xlabel("Indeks")
axs[0].set_ylabel("Nilai")

# Grafik 2: Data uniform setelah diurutkan
axs[1].plot(dis_uniform_sorted, marker='o', linestyle='-', color='purple')
axs[1].set_title("Distribusi Uniform (Setelah Diurutkan)")
axs[1].set_xlabel("Urutan (setelah sort)")
axs[1].set_ylabel("Nilai")

plt.tight_layout()
plt.show()
print(dis_uniform.mean())
```
